##**Graficando un objeto de NumPy**

##**Etiquetas**

```python
x = np.array([2, 3, 5, 7, 11, 13])
y = np.array([1, 2, 3, 4, 5, 6])

plt.plot(x, y)

plt.xlabel("Average Pulse")#("x")
plt.ylabel("Calorie Burnage")#("y")

plt.show()
```

##**Titulo**

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 3, 5, 7, 11, 13])

plt.plot(x, y)

plt.title("Nombre de la grfica")
plt.xlabel("Titulo eje x")
plt.ylabel("Titulo eje y")

plt.show()
```

##**Grid**

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 3, 5, 7, 11, 13])

plt.plot(x, y)

plt.grid(color= "blue", linestyle = "--", linewidth= 0.6 )# axis = "x", axis = "y"

plt.title("Nombre de la grfica")
plt.xlabel("Titulo eje x")
plt.ylabel("Titulo eje y")

plt.show()
```
