```python
import pandas as pd
import matplotlib.pyplot as plt

link="https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv"
code = "?raw=true"
ulr = link+code
df = pd.read_csv(ulr,sep=",", index_col=1)
```

```python
df.head()
```

```python
df.tail()
```

```python
conteo=df.groupby("Country").size().reset_index(name="Conteo")
conteo
```

```python
SL=df[df["Country"]=="Sierra Leone"]
SL
```

```python
mean_values=SL["No. of confirmed deaths"].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace=True)
```

```python
SL=SL.drop("Country", axis=1)
SL
```

```python
SL.head()
SL
```

```python
Dates=SL.index[0:-1:10]
NCC=SL["No. of confirmed cases"][0:-1:10]
NCD=SL["No. of confirmed deaths"][0:-1:10]
```

```python
fig=plt.figure(figsize=(12,6))
plt.plot(Dates,NCC,"o--r",ms=4,label="Infectados")
plt.bar(Dates,NCC,color="red",alpha=0.3)
plt.plot(Dates,NCD,"o-k",ms=4,label="Decesos")
plt.bar(Dates,NCD,color="black",alpha=0.5)
plt.grid()
plt.title("Avance del ebola en Sierra Leona")
plt.ylabel("Casos confirmados")
plt.xlabel("Periodo de tiempo")
plt.xticks(rotation=90, fontsize=8)
plt.legend()
plt.savefig("AvanceEbolaSierraLeona.pdf",dpi=300,format="pdf")
plt.show()
```

```python
 PIvsD=100*SL["No. of confirmed deaths"]/SL["No. of confirmed cases"]
 SL["PIvsD"]=PIvsD
```

```python
recuperados=((100-PIvsD)/100)*SL["No. of confirmed cases"]
SL["recuperados"]=recuperados
```

```python
SL["PR"]=(100-PIvsD)
```

```python
PIvsD_mean=SL["PIvsD"].mean()
PR_prom=SL["PR"].mean()
porcentajes=[PIvsD_mean,PR_prom]
etiquetas=["PIvsD","PR"]
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct="%0.2f %%",colors=["yellow","purple"])
plt.axis("equal")
plt.title("Porcentaje de REcuperados contra Decesos")
plt.show()
```
